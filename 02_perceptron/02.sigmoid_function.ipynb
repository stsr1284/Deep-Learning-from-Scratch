```python
# 계단 함수
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
# numpy 개선 계단 함수
import numpy as np
def step_function_np(x):
    y = x > 0
    return y.astype(np.int)
```

```python
# 위에 설명
x = np.array([-1.0, 0.0, 1.0, 2.0])
print(x)
# numpy 배열에 부등호 연산을 수행하면 bool 배열이 생긴다.
y = x > 0
print(y)

# astype를 사용하여 bool 배열을 int 배열로 변환
y = y.astype(np.int32)
print(y)
```

```python
import matplotlib.pyplot as plt
def step_function(x):
    return np.array(x > 0, dtype=np.int32)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Step Function")
plt.show()
```

## 시그모이드(Sigmoid) 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 0.0, 1.0, 2.0])
print(sigmoid(x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sigmoid Function")
plt.show()
```
