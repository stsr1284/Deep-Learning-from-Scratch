```python
import numpy as np
import matplotlib.pyplot as plt

# 0 ~ 6까지 0.1 간격으로 생성
x = np.arange(0, 6, 0.1)
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.title("Sine Wave")
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
# linestyle='--'는 점선으로 그리기
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
